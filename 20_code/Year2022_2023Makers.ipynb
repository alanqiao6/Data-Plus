```python
import pandas as pd
```

```python
# Load the datasets
MakerApplication2022 = pd.read_excel(
    "Fall2022Spring2023_MakerApplication_Scrubbed.xlsx"
)
MakerPostSurvey2022 = pd.read_excel("Fall2022Spring2023_MakerPostSurvey_Scrubbed.xlsx")
MakerPreSurvey2022 = pd.read_excel("Fall2022Spring2023_MakerPreSurvey_Scrubbed.xlsx")
```

```python
# Lets see the size of the datasets
print(MakerApplication2022.shape)
print(MakerPostSurvey2022.shape)
print(MakerPreSurvey2022.shape)
```

```python
# Merge MakerApplication2022 with MakerPreSurvey2022
MakerApplicationALL_2022 = pd.merge(
    MakerApplication2022, MakerPreSurvey2022, on="ID", how="left"
)
```

```python
# Check the size of the merged dataset
print(MakerApplicationALL_2022.shape)
```

```python
# Merge MakerApplicationALL_2022 with MakerPostSurvey2022
MakerApplicationALL_2022 = pd.merge(
    MakerApplicationALL_2022, MakerPostSurvey2022, on="ID", how="left"
)
```

```python
# Save the merged dataset as a csv file
MakerApplicationALL_2022.to_csv("MakerApplicationALL_2022.csv", index=False)
```
