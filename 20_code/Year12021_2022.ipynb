```python
import pandas as pd

# Load datasets Fall 2021 - Spring 2022
LearnerPostSurvey = pd.read_excel("Spring2022_LearnerPostSurvey_Scrubbed.xlsx")
LearnerPreSurvey = pd.read_excel("Spring2022_LearnerPreSurvey_Scrubbed.xlsx")
LearnerRegistration = pd.read_excel("Spring2022_LearnerRegistration_Scrubbed.xlsx")
LearnerRoster = pd.read_excel("Spring2022_LearnerRoster_Scrubbed.xlsx")
```

```python
# Lets see the size of the datasets
print("LearnerPostSurvey: ", LearnerPostSurvey.shape)
print("LearnerPreSurvey: ", LearnerPreSurvey.shape)
print("LearnerRegistration: ", LearnerRegistration.shape)
print("LearnerRoster: ", LearnerRoster.shape)
```

```python
LearnerRoster.head(3)
```

```python
# Merge LearnerRegistration with LearnerRoster
LearnerRegistration_merged = pd.merge(
    LearnerRegistration, LearnerRoster, on="ID", how="left"
)

# Check the size of the original datasets
print("LearnerRegistration: ", LearnerRegistration.shape)
print("LearnerRoster: ", LearnerRoster.shape)

# Check the size of the merged dataset
print("LearnerRegistration_merged: ", LearnerRegistration_merged.shape)
```

```python
LearnerRegistration_merged.head(3)
```

```python
# Check the size of the original datasets
print("LearnerRegistration_merged: ", LearnerRegistration_merged.shape)
print("LearnerPreSurvey: ", LearnerPreSurvey.shape)

# Merge LearnerRegistration_merged with LearnerPreSurvey, right on Participant
LearnerRegistration_merged = pd.merge(
    LearnerRegistration_merged, LearnerPreSurvey, on="ID", how="left"
)

# Size after the merge
print("LearnerRegistration_merged: ", LearnerRegistration_merged.shape)
```

```python
# Download the merged dataset to a csv file
LearnerRegistration_merged.to_csv("LearnerRegistration_Year1_2021_2022.csv", index=False)
```
