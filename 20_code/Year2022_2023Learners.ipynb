```python
import pandas as pd
```

```python
# Load datasets Learners Spring 2023
LearnerPostSurvey2023 = pd.read_excel("Spring2023_LearnerPostSurveyAll_Scrubbed.xlsx")
LearnerPreSurveyAll2023 = pd.read_excel("Spring2023_LearnerPreSurveyAll_Scrubbed.xlsx")
LearnerRegistration2023 = pd.read_excel("Spring2023_LearnerRegistration_Scrubbed.xlsx")
LearnerRoster_Scrubbed2023 = pd.read_excel("Spring2023_LearnerRoster_Scrubbed.xlsx")
```

```python
# Lets see the size of the datasets
print(LearnerPostSurvey2023.shape)
print(LearnerPreSurveyAll2023.shape)
print(LearnerRegistration2023.shape)
print(LearnerRoster_Scrubbed2023.shape)
```

```python
# Merge LearnerRegistration with LearnerRoster
LearnerRegistration_merged = pd.merge(
    LearnerRegistration2023, LearnerRoster_Scrubbed2023, on="ID", how="left"
)

# Check the size of the original datasets
print("LearnerRegistration: ", LearnerRegistration2023.shape)
print("LearnerRoster: ", LearnerRoster_Scrubbed2023.shape)

# Check the size of the merged dataset
print("LearnerRegistration_merged: ", LearnerRegistration_merged.shape)
```

```python
LearnerRegistration_merged.head(3)
```

```python
# Check the size of the original datasets
print("LearnerRegistration_merged: ", LearnerRegistration_merged.shape)
print("LearnerPreSurvey: ", LearnerPreSurveyAll2023.shape)

# Merge LearnerRegistration_merged with LearnerPreSurvey, right on Participant
LearnerRegistration_merged = pd.merge(
    LearnerRegistration_merged, LearnerPreSurveyAll2023, on="ID", how="left"
)

# Size after the merge
print("LearnerRegistration_merged: ", LearnerRegistration_merged.shape)
```

```python
# Check the size of the original datasets
print("LearnerRegistration_merged: ", LearnerRegistration_merged.shape)
print("LearnerPostSurvey: ", LearnerPostSurvey2023.shape)

# Merge LearnerRegistration_merged with LearnerPostSurvey, right on Participant
LearnerRegistration_merged = pd.merge(
    LearnerRegistration_merged, LearnerPostSurvey2023, on="ID", how="left",suffixes=('_left', '_right')
)

# Size after the merge
#print("LearnerRegistration_merged: ", LearnerRegistration_merged.shape)
```

```python
# Download the merged dataset to a csv file
LearnerRegistration_merged.to_csv(
    "LearnerRegistration_Year2_2022_2023.csv", index=False
)
```
